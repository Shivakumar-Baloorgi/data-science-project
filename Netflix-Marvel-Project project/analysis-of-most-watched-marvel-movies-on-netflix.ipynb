```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/datasets/shivakumarbaloorgi/netflix-marvel-movies-dataset/marvel_netflix_movies.csv.txt')

df.head()
```

```python
print(df.info())

print(df.describe())
```

```python
top_movie = df .sort_values(by='Views_Million', ascending=False)

print(top_movie)
```

```python
sns.barplot(
    x='Views_Million',
    y='Movie',
    hue='Movie',
    data=top_movie,
    palette='Reds_r',
    legend=False
)

plt.title('Most Watched Marvel Movies on Netflix')

plt.xlabel('Views in Millions')

plt.ylabel('Movie')

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x='IMDb_Rating',
    y='Views_Million',
    hue='Movie',
    s=200,
    data=df
)

plt.title('IMDb Rating vs Views')

plt.show()
```

```python
correlation = df['IMDb_Rating'].corr(df['Views_Million'])

print("Correlation:", correlation)
```

# Insights

1. Spider-Verse has highest views and ratings.
2. Higher IMDb ratings generally lead to more views.
3. Some movies got high views despite lower ratings.
